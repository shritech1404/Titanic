## Drop unuseble attributes and print again

```python
inputs = a.drop(['PassengerId','Survived', 'Name', 'SibSp', 'Parch', 'Parch', 'Cabin', 'Embarked'], axis= 'columns')
```

```python
inputs
```

```python
inputs.head()
```

## Print our targeted attribute
* our goal is to predict the passenger is Servived or not, so that's why our targeted attribute is Survived

```python
target = a['Survived']
```

```python
target
```

## convert all atrributes those who are in string convert into number by using LabelEncoder method

```python
from sklearn.preprocessing import LabelEncoder
```

## Encode the attributes

```python
le_Sex = LabelEncoder()
le_Age = LabelEncoder()
le_Ticket = LabelEncoder()
le_Fare = LabelEncoder()
```

## Create new attributes and store the Encoded attributes in new attributes and print again dataset

```python
inputs['Sex_n'] = le_Sex.fit_transform(inputs['Sex'])
inputs['Age_n'] = le_Age.fit_transform(inputs['Age'])
inputs['Ticket_n'] = le_Ticket.fit_transform(inputs['Ticket'])
inputs['Fare_n'] = le_Fare.fit_transform(inputs['Fare'])
```

```python
inputs
```

## Now our numbers attributes are printed so we are drop the string attributes and print it.

```python
inputs_n = inputs.drop(['Sex','Age','Ticket','Fare'], axis='columns')
```

```python
inputs_n
```

## now we are going to train our model by using dicision treee.

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

## fit our model

```python
model.fit(inputs_n,target)
```

## find the score of our model

```python
model.score(inputs_n,target)
```

## now our model is ready to predict

```python
model.predict([[3,1,28,523,18]])
```

## [3,1,28,523,18] this is the first attributes in the training dataset and the predicted output is 0 passenger is not servived

## now we are going to check our test data

```python
model.predict([[2,0,20,52,144]])
```

## [2,0,20,52,144] this is the test attributes and the predicted output is 1 passenger is  servived
